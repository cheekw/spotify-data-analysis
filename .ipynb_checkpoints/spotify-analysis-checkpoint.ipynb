# Spotify Data Analysis

```python
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path = "./data/SpotifyFeatures.csv"
df = pd.read_csv(data_path)
```

```python
df.corr()
```

```python
fig = plt.figure(figsize = (20, 15))
ax = fig.gca()
df.hist(ax = ax, bins=20)
```

```python
category_mean_df = df.groupby(['genre']).mean().reset_index()
```

```python
fig = plt.figure(figsize = (27, 10))
sns.boxplot(x="genre", y="energy", data=df)
```

```python
fig = plt.figure(figsize = (27, 10))
sns.boxplot(x="genre", y="popularity", data=df)
```

```python
fig = plt.figure(figsize = (27, 10))
sns.boxplot(x="genre", y="danceability", data=df)
```
